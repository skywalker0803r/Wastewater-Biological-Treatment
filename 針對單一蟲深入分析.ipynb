```python
import pandas as pd
from tqdm import tqdm_notebook as tqdm

df = pd.read_csv('/content/drive/MyDrive/廢水案子/ALLdatLen.csv')
df
```

```python
d = df.loc[df.cName=='太陽蟲']
d
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
d['D'] = ss.fit_transform(d[['D']])
d
```

```python
import matplotlib.pyplot as plt
plt.boxplot(d['D'])
```

# drop outlier

```python
m = d['D'].mean()
ub = d['D']<= m + (3*d['D'].std())
lb = d['D']>= m - (3*d['D'].std())
cond = ub & lb
d = d.loc[cond]
plt.boxplot(d['D'])
```

# linear regression

```python
from sklearn.model_selection import train_test_split
import numpy as np

X = [i for i in range(len(d['date']))]
y = d['D']
X = np.array(X).reshape(-1,1)
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

# Autoregression

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 拟合模型
model = sm.OLS(endog=y_train, exog=X_train).fit()
# 打印模型摘要
print(model.summary())

# 绘制原始数据和预测值的对比图
plt.figure(figsize=(12, 6))
plt.scatter(X_train, y_train, label='Original Data', color='blue')
plt.scatter(X_train, model.predict(), label='AR(1) Model', color='red')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
