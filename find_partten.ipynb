```python
import pandas as pd
from tqdm import tqdm_notebook as tqdm
from datetime import datetime

df = pd.read_csv('/content/drive/MyDrive/廢水案子/ALLdatLen.csv')
df['date'] = pd.to_datetime(df['date'],format="%Y%m%d")
df
```

```python
df.loc[df.cName=='太陽蟲']
```

```python
import plotly.graph_objs as go

trace = go.Scatter(
    x=df.loc[df.cName=='太陽蟲','date'],
    y=df.loc[df.cName=='太陽蟲','D'],
    mode='lines+markers')

# 創建圖表佈局
layout = go.Layout(title='日期 vs. 數值',
                   xaxis=dict(title='日期'),
                   yaxis=dict(title='數值'))

# 創建圖表物件
fig = go.Figure(data=[trace], layout=layout)

# 顯示圖表
fig.show()
```

```python
for i in df.cName.unique():
  trace = go.Scatter(
    x=df.loc[df.cName==i,'date'][:100],
    y=df.loc[df.cName==i,'D'][:100],
    mode='lines+markers')

  # 創建圖表佈局
  layout = go.Layout(title=i,xaxis=dict(title='日期'),yaxis=dict(title='數值'))

  # 創建圖表物件
  fig = go.Figure(data=[trace], layout=layout)

  # 顯示圖表
  fig.show()
```
